                                  İngiltere, Leeds Şehri İçin Hava Durumu İncelemesi

Leeds Şehrinin Hava Durumu Hakkında Kısa Bilgilendirme:

Leeds, İngiltere'nin kuzeyinde bulunan güzel bir şehirdir. Genellikle ılıman bir iklimi vardır. Yaz ayları genellikle serin ve yağışlı olabilir, sıcaklıklar 15-20°C arasında değişebilir. Kış ayları ise soğuk ve nemli olabilir, sıcaklıklar 0-5°C arasında seyreder. Leeds genellikle yağışlı bir şehir olduğundan, şemsiyenizi yanınıza almak iyi bir fikir olabilir. Şehirde bulunan yeşil alanlar ve tarihi mekanlar, ziyaretçilere keyifli bir deneyim sunar. Unutmayın ki İngiltere'nin hava koşulları değişken olabilir, bu yüzden seyahat ederken hava durumu güncellemelerini takip etmek önemlidir.

Şimdi Leeds şehrinin 2006 başı ile 2017 başına kadar olan aralıkta ki hava durumu verilerinim istatstiklerine bakalım:
    

```python
import matplotlib.pyplot as plt


etiketler = ['Parçalı Bulutlu', 'Çok Bulutlu', 'Diğer']
yuzdeler = [33, 29, 38]
renkler = ['#ff9999', '#66b3ff', '#99ff99']
explode = (0.1, 0, 0)  # Sadece "Parçalı Bulutlu" dilimi patlatır


plt.figure(figsize=(8, 8))
plt.pie(yuzdeler, labels=etiketler, autopct='%1.1f%%', startangle=90, colors=renkler, explode=explode)
plt.title('Parçalı Bulutlu Veri Dağılımı')


plt.axis('equal') 
plt.show()
```

```python
import matplotlib.pyplot as plt


etiketler = ['Yağmur', 'Kar', 'Diğer']
yuzdeler = [88, 11, 1]
renkler = ['#ff9999', '#66b3ff', '#99ff99']
explode = (0.1, 0, 0) 


plt.figure(figsize=(8, 8))
plt.pie(yuzdeler, labels=etiketler, autopct='%1.1f%%', startangle=90, colors=renkler, explode=explode)
plt.title('Yağış Türü Dağılımı')


plt.axis('equal')  
plt.show()
```

```python
import matplotlib.pyplot as plt


veriler = [-21.8, 4.69, 12, 18.8, 39.9]


plt.figure(figsize=(8, 6))
plt.boxplot(veriler, vert=False, widths=0.7, patch_artist=True)
plt.xlabel('Sıcaklık Verileri')
plt.title('Sıcaklık Verileri Box Plot')
plt.xticks(fontsize=12)


plt.tight_layout()
plt.show()
```
